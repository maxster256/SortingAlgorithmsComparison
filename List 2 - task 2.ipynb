```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

qs = pd.read_csv("tests_quicksort.csv")
ms = pd.read_csv("tests_merge.csv")
ins = pd.read_csv("tests_insert.csv")

qs.columns = ["Iteration", "Trial", "Size", "Comparisons", "Swaps", "Duration (s)"]
ms.columns = ["Iteration", "Trial", "Size", "Comparisons", "Swaps", "Duration (s)"]
ins.columns = ["Iteration", "Trial", "Size", "Comparisons", "Swaps", "Duration (s)"]

qs.head(3)
```

```python
frames = [qs, ms, ins]
avg = pd.concat(frames)

avg.groupby('Size', as_index=False)['Comparisons'].mean()
```

```python
avg.groupby('Size', as_index=False)['Swaps'].mean()
```

```python
avg.groupby('Size', as_index=False)['Duration (s)'].mean()
```

```python
def cn(df_sub):
    return df_sub['Comparisons'].sum() / float(df_sub['Size'].sum())

avg.groupby('Size').apply(cn)
```

```python
def sn(df_sub):
    return df_sub['Swaps'].sum() / float(df_sub['Size'].sum())

avg.groupby('Size').apply(sn)
```

```python
qs.plot(x='Size', y='Swaps')
qs.plot(x='Size', y='Comparisons')
qs.plot(x='Size', y='Duration (s)')
```

```python
fig = plt.figure()


plt.plot(qs['Size'], qs['Comparisons'], label="Quick")
plt.plot(ms['Size'], ms['Comparisons'], label="Merge")
plt.plot(ins['Size'], ins['Comparisons'], label="Insert")

plt.title("Combined Statistics")
plt.xlabel("Total comparisons")
plt.ylabel("Size of data (n)")

plt.legend(loc='upper left')
plt.show()
```

```python
fig = plt.figure()

for frame in [qs, ms]:
    plt.plot(frame['Size'], frame['Swaps'])

plt.title("Combined Statistics")
plt.xlabel("Total swaps")
plt.ylabel("Size of data (n)")
plt.show()
```

```python
fig = plt.figure()

for frame in [qs, ms, ins]:
    plt.plot(frame['Size'], frame['Duration (s)'])

plt.title("Combined Statistics")
plt.xlabel("Total duration (s)")
plt.ylabel("Size of data (n)")
plt.show()
```
